#1. Wprowadzenie
  W tym kursie nauczysz się o sztucznej inteligencji w języku Python, ale najpierw musimy wyjaśnić kilka pojęć, aby sklaryfikować pewnych nieścisłości.
 Na wstępie warto wyjaśnić różnice pomiędzy sztuczną inteligencją a uczeniem maszynowym.

Sztuczna inteligencja jest inteligencją hipotetyczną - Realizowana ona jest w procesie technicznym, a nie naturalnym. Zadaniem jej jest naśladowanie inteligencji ludzkiej poprzez analizę i przetwarzanie informacji. Sztuczna inteligencja sama w sobie jest realizowana za pomocą określonych schematów matematycznych, których zadaniem będzie wypracowanie metody na rozwiązanie danego problemu

![grafika1](https://drive.google.com/uc?id=1B5b9LoTDyppCWTHrKiqauo37vRp8FqH6)

  Jedną z dziedzin SI jest uczenie maszynowe (machine learning). W normalnym przypadku, kiedy mamy dane, oraz algorytm, otrzymujemy wynik. Tutaj jednak pojawia się najbardziej oczekiwany moment przez osoby pragnące zapoznać sie z SI. Można zlecić maszynie wypracowanie algorymu, poprzez podanie danych oraz oczekiwanego wyniku. Odpowiednio przygotowany model matematyczny używa danych do opracowania metody, według której otrzymano wyniki. Wytrenowany w ten sposób model jest później w stanie samemu wypracować wyniki zbliżone rzeczywistym dla danych, których jeszcze nie poznał.  Sposób na przygotowanie takiego modelu zostanie przedstawiony w dalszej części kursu.
  
  Na uczenie maszynowe składa się między innymi uczenie głebokie (deep learning) oraz sieci neuronowe (neural networks), o których można będzie uzyskać 
więcej informacji w dalszej części kursu.

#2. Zastosowanie
Warto porozmawiać o zastosowaniach sztucznej inteligencji, których nie można aktualnie zliczyć. Dzieje się to ze względu na ciągłe odnajdywanie nowych sposobów automatyzacji w różnych branżach. Możemy sobie nie zdawać z tego sprawy ale sztuczna inteligencja otacza nas występując w wielu aspektach codziennego życia. 

###2.1. Google
Każdy z nas używa na codzień internetu, czy do szukania informacji, pracy, rozrywki, zastosowań jest wiele. Nie ma chyba osoby która przy próbie zalogowania się do różnych portali nie została poproszona o wskazanie zdjęć zawierających konkretne elementy. Jest to [CAPTCHA](https://support.google.com/a/answer/1217728?hl=pl), system firmy google który służy do weryfikacji, czy użytkownik nie jest botem. Instnieją pogłoski że zdjęcia rozpoznawane przez ludzi są używane do trenowania sztucznej inteligencji.

Wiele osób słusznie zauważa, że wraz z przeglądaniem stron internetowych, reklamy, które widzimy na stronach pokazują rzeczy, które zwykle wyszukujemy. Reklamy te są często elementem systemu firmy Google, która używa SI w wielu fascynujących projektach o których możemy dowiedzieć się wiecej adresem https://blog.google/technology/ai/.

###2.2. Tesla
Firma ta jest znana w branży SI za sprawą autonomicznych aut elektrycznych, posiadające autopilota który sam prowadzi auto do wskazanego celu. Ważnym elementem tego produktu są systemy bezpieczeństwa, które muszą błyskawicznie reagować na to co dzieje się na drodze. System ten jest ciągle dopracowywany pod względem technicznym. Sztuczna inteligencja sterująca autonomicznym trybem jazdy również jest udoskonalana dzięki danym które zbierają auta. Więcej informacji można znaleść pod adresem https://www.tesla.com/AI.

###2.3. Boston Dynamics
Jest znaną firmą od lat zajmującą się tworzeniem i udoskonalaniem robotów. Często z tą firmą jest kojarzony robot [Spot](https://www.vision-systems.com/embedded/article/14179537/artificial-intelligence-software-expands-boston-dynamics-spot-robot-capabilities).
![Spot](https://drive.google.com/uc?id=1CkTt5VfiHfMxj-88zQf14taDkDoPdkGC)

Robot ten jest efektem wieloletniej pracy wybitnych inzynierów, którzy spędzili setki godzin w celu usprawnienia robota, oraz SI, która jest używana do reagowania na otoczenie, w którym się znajduje. Firma ta posiada [inne modele](https://general-laser.at/en/products-en#ROBOTS) robotów posiadające ogromny potencjał do wykorzystania.



Są to tylko przykłady kilku firm znanych ze swoich osiągnieć w branży SI pod względem komercyjnym. Istnieje jednak wiele innych ambitnych projektów, którym warto się przyjrzeć. 

###2.4. Youtube

Jest to najpopularniejsza platforma internetowa na świecie służąca do publikowania zawartości audio/video. Zawiera niezliczoną ilość godzin zawartości w prawie każdym temacie, jest ona również popularnym miejscem do umieszczania tutoriali w różnej tematyce. Listę ciekawych kanałów/filmów realizujących materiały o tematyce SI w formie prostych projektów można znaleść w sekcji [**Odnośniki**](#scrollTo=NtrDIlkCEMsa)

Wypisanie [źródeł](#scrollTo=NtrDIlkCEMsa), które dotyczą każdego wykorzystania SI w obecnych czasach jest niemożliwe do wykonania w realnym czasie, dlatego zachęca się uczestnika do szukania dodatkowych infromacji na własną rękę. Samo wpisanie pojęcia "Sztuczna inteligencja" wraz z tematem który nas interesuje w google, umożliwi otrzymanie ogromnej ilości fascynujących materiałów. To tylko dowodzi tego jak rozległe i ciekawe jest zagadnienie sztucznej inteligencji.
